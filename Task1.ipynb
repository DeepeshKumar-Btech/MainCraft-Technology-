# Student Performance Analysis

This project analyzes student performance using Python, pandas, and matplotlib.  
The dataset contains information about student grades, study time, gender, and other factors affecting academic performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading the Dataset

The student performance dataset is loaded into a pandas DataFrame using the read_csv() function.
# Displaying Dataset

The first few rows of the dataset are displayed to understand its structure and contents.

```python
df = pd.read_csv("student-mat.csv", sep=';')

print(df.head())
```

# Checking Dataset Shape

This step shows the number of rows and columns present in the dataset.

```python
print("\nDataset Shape:")
print(df.shape)
```

# Checking Data Types

The datatype of each column is checked to understand whether the data is numerical or categorical.

```python
print("\nData Types:")
print(df.dtypes)
```

# Checking Missing Values

This step checks whether the dataset contains any missing or null values.

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

# Removing Duplicate Rows

Duplicate rows are removed from the dataset to improve data quality and analysis accuracy.

```python
df = df.drop_duplicates()

print("\nShape After Removing Duplicates:")
print(df.shape)
```

# Average Final Grade

Calculating the average final grade (G3) of students in the dataset.

```python
average_grade = df['G3'].mean()

print("\nAverage Final Grade (G3):")
print(average_grade)
```

# Students Scoring Above 15

Finding the number of students who scored more than 15 in the final grade.

```python
high_scores = df[df['G3'] > 15]

print("\nNumber of Students Scoring Above 15:")
print(len(high_scores))
```

# Correlation Between Study Time and Performance

This step checks whether study time has any relationship with student final grades.

```python
correlation = df['studytime'].corr(df['G3'])

print("\nCorrelation Between Study Time and Final Grade:")
print(correlation)
```

# Gender-wise Average Performance

Comparing the average final grades of male and female students.

```python
gender_avg = df.groupby('sex')['G3'].mean()

print("\nAverage Final Grade by Gender:")
print(gender_avg)
```

# Data Visualization

The following graphs help visualize student performance trends and relationships in the dataset.
# Histogram of Final Grades

A histogram is plotted to visualize the distribution of students' final grades.

```python
plt.hist(df['G3'], bins=10)

plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")

plt.show()
```

# Scatter Plot: Study Time vs Final Grade

This scatter plot helps visualize the relationship between study time and student performance.

```python
plt.scatter(df['studytime'], df['G3'])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

# Bar Chart: Male vs Female Average Scores

This bar chart compares the average final grades of male and female students.

```python
gender_avg.plot(kind='bar')

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")

plt.show()
```

# Conclusion

- The dataset was successfully loaded and analyzed.
- Missing values and duplicate rows were checked and cleaned.
- Average student performance was calculated.
- Correlation between study time and grades was analyzed.
- Gender-wise performance comparison was completed.
- Visualizations helped understand performance trends clearly.
